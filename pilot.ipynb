```python
import networkx as nx
import matplotlib.pyplot as plt

def create_background_network(num_nodes=200, num_edges=300, area_extent=10):
    G_background = nx.gnm_random_graph(num_nodes, num_edges)
    pos_background = nx.spring_layout(G_background, scale=area_extent)
    return G_background, pos_background

def integrate_text_with_network(text_graph, text_pos, background_graph, background_pos, shift_factor=(0, -2)):
    combined_graph = nx.compose(text_graph, background_graph)
    combined_pos = {}
    combined_pos.update(background_pos)  # Add background positions
    
    for node, position in text_pos.items():
        shifted_pos = (position[0] + shift_factor[0], position[1] + shift_factor[1])
        combined_pos[node] = shifted_pos

    return combined_graph, combined_pos

def draw_combined_network(graph, pos, text_graph, text_pos):
    plt.figure(figsize=(14, 8))
    nx.draw(graph, pos, node_size=20, node_color='gray', edge_color='lightgray', alpha=0.4, with_labels=False)
    nx.draw(text_graph, text_pos, node_size=80, node_color='black', edge_color='black', alpha=1.0, with_labels=False)
    plt.show()

def export_to_graphml(graph, pos, filename="network.graphml"):
    # Add position attributes to nodes
    for node, position in pos.items():
        graph.nodes[node]['x'] = position[0]
        graph.nodes[node]['y'] = position[1]
    
    # Export graph to GraphML format
    nx.write_graphml(graph, filename)
    print(f"Graph exported to {filename}")

def export_text_graph(text_graph, text_pos, filename="text_graph.graphml"):
    # Add position attributes to nodes for the text graph
    for node, position in text_pos.items():
        text_graph.nodes[node]['x'] = position[0]
        text_graph.nodes[node]['y'] = position[1]
    
    # Export the text graph only to GraphML format
    nx.write_graphml(text_graph, filename)
    print(f"Text graph exported to {filename}")

def export_background_graph(background_graph, background_pos, filename="background_graph.graphml"):
    # Add position attributes to nodes for the background graph
    for node, position in background_pos.items():
        background_graph.nodes[node]['x'] = position[0]
        background_graph.nodes[node]['y'] = position[1]
    
    # Export the background graph only to GraphML format
    nx.write_graphml(background_graph, filename)
    print(f"Background graph exported to {filename}")

# Example Usage
text = "STATE OF THE INTERACTOMES"
text_graph, text_pos = text_to_graph(text)
background_graph, background_pos = create_background_network()
combined_graph, combined_pos = integrate_text_with_network(text_graph, text_pos, background_graph, background_pos, shift_factor=(3, -2))

# Visualize the combined network
#draw_combined_network(combined_graph, combined_pos, text_graph, text_pos)

# Export the combined network to GraphML for Cytoscape
#export_to_graphml(combined_graph, combined_pos, filename="network_with_text.graphml")

# Export the text graph only to GraphML
export_text_graph(text_graph, text_pos, filename="/cellar/users/snwright/Data/Transfer/pcnet/text_graph.graphml")

# Export the background graph only to GraphML
export_background_graph(background_graph, background_pos, filename="/cellar/users/snwright/Data/Transfer/pcnet/background_graph.graphml")

```
